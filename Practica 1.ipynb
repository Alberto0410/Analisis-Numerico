## Ejercicio 1
El número de Euler se puede definir como:
## $$ e =\underset{n\rightarrow\infty}{\lim} \left(1 + \dfrac{1}{n}\right)^n$$</font>
Escribe un programa que, a manera de tabla arroje los valores de $(1 + 1/n)^n$ para $n = 10^k$, con $k = 0,1,2,...,20$, y en cada paso calcule los errores absolutos y relativos comparando con el valor de $e$ dado por la máquina. Utiliza doble precisión de IEEE para hacer los cálculos.

```python
from math import e
import pandas as pd
import decimal
import math
import numpy as np

#definimos una funcion que calcule el limite
def exp(n):
  return((1+ (1/n))**n)

#creamos las listas que van a tener los valores absolutos y realtivos
error_absoluto = []
error_relativo = []
exponente = []

#hacemos las 20 iteraciones y guardamos los resultado
for i in range(20):
  aprox = exp(10**i)
  error_r = abs(e - aprox)
  error_a = error_r/abs(e)

  #usamos la libreria decimal para guardar todos los decimales
  error_absoluto.append(decimal.Decimal.from_float(error_a))
  error_relativo.append(decimal.Decimal.from_float(error_r))
  exponente.append('10^' + str(i+1))

#Creamos un dataframe para mostrar los resultados
errores = pd.DataFrame({'Valor de n' : exponente, 'Error relativo': error_relativo, 'Error absoluto' : error_absoluto})

errores
```

```python
print(len(str(decimal.Decimal.from_float(exp(10**14)))))
```

> - <font color=red> ¿Qué observas con las aproximaciónes de $e$ y los errores?</font>

Conforme el exponente de $10^n$ aumenta el error se va reduciendo hasta que llega a $n = 13$ que es cuando el error aumenta, es decir que el error de la aproximación $n = 13$ es mayor que el de la aproximación $n = 14$
> - <font color=red> Explica lo que sucede.</font>

Lo anterior se puede explicar ya que cuando el exponente llega a $n = 13$ la aproximación tiene una longitud de 51 dígitos, lo cual es casi la misma cantidad de bits disponibles para la mantisa. Luego, cuando $n = 14$ la aproximación supera el número de bits disponibles para la mantisa, con lo cual se empieza a truncar/redondear por lo que vamos a generar errores en el proceso, lo cual explica lo sucedido

## Ejercicio 2
La función exponencial se puede aproximar con su serie de Taylor alrededor de cero: 

## $$e^{x} = 1 + x + \frac{x^{2}}{2} + \cdots + \frac{x^{n}}{n!} + \cdots $$

Escribe un programa que calcule el valor aproximado de $e^x$ para cualquier $x$ mediante los primeros $n$ términos de su serie de Taylor. Como criterio o condición de paro propón que $\frac{x^n}{n!}<10^{-6}$. El programa debe evitar cancelaciones para valores negativos de x.

```python
def exp_pos(x, n):
  exp = 0

  for i in range (n):
      if (x**i)/math.factorial(i) < 10**(-6):
        break
        
      exp += decimal.Decimal.from_float(((x**i)/math.factorial(i)))
  return exp

def expo(x, n):
  if  x > 0:
    return exp_pos(x, n)
  else:
    return(1/(exp_pos(-x, n)))
```

## Ejercicio 3

Investiga como obtener la épsilon de máquina $\varepsilon$, en doble precisión, de Python. Implementa lo anterior en un programa e imprime el valor de $\varepsilon$, además calcula la siguiente cantidad:

## $$\epsilon = \vert 3(4/3 -1) -1 \vert$$

Imprime el valor de $\epsilon$.

```python
import numpy as np

abs_eps = abs(3*(4/3-1)-1)

def epsilonMaq(func=float):
  maq_epsilon = func(1)
  while func(1)+func(maq_epsilon) != func(1):
    maq_epsilon_n = maq_epsilon
    maq_epsilon = func(maq_epsilon) / func(2)
  return maq_epsilon_n

eps = np.finfo(np.float64).eps

emaq = epsilonMaq(np.float64)

print("ϵ = |3(4/3-1)-1| = ",abs_eps, " - - operación dada")
print("ε = ", emaq , " - - función implementada para calcular la épsilon de máquina")
print("ε_2 = ",eps, " - - épsilon calculada con finfo de numpy")
```

> - <font color=red> Explica porque $\epsilon = \varepsilon$</font>

Por definición sabemos que la épsilon de la máquina es el número decimal más pequeño que sumado a 1 nos arroja un valor mayor a 1. Vemos que $\epsilon$ es tan pequeña como $\varepsilon$ ya que al redondear $(4/3)-1 = 1/3 = 0.33\bar{3}$) y entonces tendremos que $\epsilon = \vert 3(0.33̄\bar{3}) -1 \vert = \vert 0.99\bar{9} - 1 \vert$ obteniendo así un número tan pequeño que cumple ser $ɛ$ ya que sabemos que 1/3 es "periodico" en nuestro sistema de doble precisón, así podemos asegurar que existirá un "residuo".

No funcionaría, ya que 1/3 en un sistema de punto flotante de base 3 tendría una representadión finita, y no existiría este "residuo" pequeño
